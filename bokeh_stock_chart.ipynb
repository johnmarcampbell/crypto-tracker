```python
from iexfinance import get_historical_data
from datetime import datetime
from charts import get_product_data, make_rate_plot
from bokeh.io import output_notebook, show
output_notebook()
```

```python
start = datetime(2017, 2, 9)
end = datetime(2017, 5, 24)

df = get_historical_data("GOOG", start=start, end=end, output_format='pandas')
df.reset_index(level=0, inplace=True)
df.rename(columns={'date': 'time'}, inplace=True)
df['time'] = df['time'].apply(lambda x: int(datetime.strptime(x, '%Y-%m-%d').strftime('%s')))
df['time'] = df['time'] * 1000
```

```python
fig = make_rate_plot(df)
show(fig)
```
